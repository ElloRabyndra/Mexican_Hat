# Mexican Hat — Evolusi Pola Aktivasi
Jaringan Kompetisi dengan Algoritma Mexican Hat  
**Parameter:** R1=1, R2=3, c1=0.6, c2=−0.4  
**Input awal:** x = (0.0, 0.5, 0.8, 1.0, 0.8, 0.5, 0.0)

```python
import numpy as np
import matplotlib
import matplotlib.pyplot as plt
import matplotlib.ticker as ticker

# ── Parameter ──────────────────────────────────────────────────────────────
R1 = 1          # radius bobot positif
R2 = 3          # radius bobot negatif
c1 =  0.6       # bobot positif
c2 = -0.4       # bobot negatif
n  =  7         # jumlah neuron

# ── Nilai awal (t=0) ───────────────────────────────────────────────────────
x0 = np.array([0.0, 0.5, 0.8, 1.0, 0.8, 0.5, 0.0])

# ── Fungsi bobot Mexican Hat ────────────────────────────────────────────────
def get_weight(i, j):
    """Bobot dari neuron j terhadap neuron i (0-indexed)."""
    d = abs(i - j)
    if d == 0:
        return c1          # self-connection (jarak 0 ≤ R1)
    elif d <= R1:
        return c1
    elif d <= R2:
        return c2
    else:
        return 0.0

# ── Satu langkah iterasi ────────────────────────────────────────────────────
def iterate(x):
    """Hitung satu iterasi: raw → normalisasi ke [0, x_max]."""
    raw = np.array([
        sum(get_weight(i, j) * x[j] for j in range(n))
        for i in range(n)
    ])
    x_max = raw.max()
    activated = np.clip(raw, 0, x_max)   # fungsi aktivasi: min(x_max, max(0, raw))
    return activated

# ── Jalankan 2 iterasi ──────────────────────────────────────────────────────
x1 = iterate(x0)   # t=1
x2 = iterate(x1)   # t=2

print("t=0:", np.round(x0, 4))
print("t=1:", np.round(x1, 4))
print("t=2:", np.round(x2, 4))
```

```python
# ── Plot ────────────────────────────────────────────────────────────────────
labels = [f'$x_{i+1}$' for i in range(n)]
x_pos  = np.arange(n)

fig, ax = plt.subplots(figsize=(11, 6))
fig.patch.set_facecolor('white')

# Garis & marker — warna sesuai gambar referensi
ax.plot(x_pos, x0, color='#4472C4', linewidth=2.2, marker='o', markersize=5, label='t=0')
ax.plot(x_pos, x1, color='#C0392B', linewidth=2.2, marker='o', markersize=5, label='t=1')
ax.plot(x_pos, x2, color='#7F9A2B', linewidth=2.2, marker='o', markersize=5, label='t=2')

# Judul & label
ax.set_title('Mexican Hat — Evolusi Pola Aktivasi',
             fontsize=16, fontweight='bold', pad=14)
ax.set_xlabel('Indeks Neuron', fontsize=12, labelpad=8)
ax.set_ylabel('Nilai Aktivasi', fontsize=12, labelpad=8)

# Sumbu X: ganti angka dengan label xₙ
ax.set_xticks(x_pos)
ax.set_xticklabels(labels, fontsize=11)

# Sumbu Y: format 4 desimal koma (gaya Indonesia: titik → koma)
ax.yaxis.set_major_formatter(
    ticker.FuncFormatter(lambda v, _: f'{v:.4f}'.replace('.', ','))
)
ax.set_ylim(-0.02, max(x2.max(), x1.max(), x0.max()) * 1.12)

# Grid & spines
ax.grid(axis='y', linestyle='-', linewidth=0.6, color='#CCCCCC')
ax.set_axisbelow(True)
for spine in ax.spines.values():
    spine.set_edgecolor('#AAAAAA')

# Legenda di luar area plot (kanan)
ax.legend(loc='center left', bbox_to_anchor=(1.01, 0.5),
          fontsize=11, frameon=True, edgecolor='#AAAAAA')

plt.tight_layout()
plt.savefig('mexican_hat_plot.png', dpi=150, bbox_inches='tight')
plt.show()
print("Grafik disimpan sebagai 'mexican_hat_plot.png'")
```
